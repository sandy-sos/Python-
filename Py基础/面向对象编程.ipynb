```python
#类中的属性和方法
#实例方法管“对象”，类方法管“类”，静态方法只管“工具函数”
class person():
    #定义属性
    name = '小明'
    age = 18
    #定义实例方法：在类中定义函数，且自带参数self
    def say(self):
        print('大家好，我是%s，今年%d岁了' % (self.name, self.age))

    #实例属性:调用实例方法时，必须先创建类的实例对象，然后通过实例对象调用方法。
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    #类方法:不能调用实例属性和方法，只能调用类属性和类方法。
    # 类方法的第一个参数是cls，表示类本身。
    @classmethod
    def class_method(cls):
         print('这是一个类方法')
         
    
    #静态方法:不能调用实例属性和方法，也不能调用类属性和类方法。
    @staticmethod
    def add(x, y):
        res = x + y
        return res
    


def say():
    print("Hey~~")


person = person("Andy", 11)
person.say()
person.class_method()
person.add(1, 2)
person.say=say() #函数赋值：say函数赋值给person.say，person.say()就可以调用say函数

```

```python
#封装 访问限制：单下划线开头（protected）表示受保护的属性或方法，
#双下划线开头（private）表示私有属性或方法，外部无法直接访问。
#双下划线开头的属性或方法，在类的外部无法直接访问，但可以通过类的实例对象访问。

class pet:
        def __init__(self, name, age):
            self._name = name  #受保护的
            self.__age = age    # 私有属性

pet = pet("小白", 2)
print(pet._name)  #可以访问受保护的属性
# print(pet.__age)  #无法访问私有属性，会报错

print(pet._pet__age)  #可以通过类的实例对象访问私有属性 但不安全
dir(pet)  #查看类的属性和方法，私有属性会被改名为 _类名__属性名

```

```python
#类的继承：一个子类可以继承多个父类；一个父亲可以有多个子类；
# 子类可以继承父类的属性和方法，也可以重写父类的方法。

class person():#定义一个person类默认继承object类
    def __init__(self, name, age):
        self.name = name 
        self.age = age

    def say(self):
        print('大家好，我是%s，今年%d岁了' % (self.name, self.age))

class student(person):#定义一个student类继承person类
    def __init__(self, name, age, grade):
        super().__init__(name, age) #调用父类的构造方法
        self.grade = grade

    def say(self):#重写父类的方法 类似就近原则
        print('大家好，我是%s，今年%d岁了，我是%s年级的学生' % (self.name, self.age, self.grade))
        
```

```python
#多继承
class A():
    def printA(self):
        print("A类")
class B():
    def printB(self):
        print("B类")    
class C(A, B):#C类继承A类和B类
    def printC(self):
        print("C类")
c1=C()#创建C类的实例对象
c1.printA()#调用A类的方法
c1.printB()#调用B类的方法   
c1.printC()#调用C类的方法
```

```python
#多态：同一个方法在不同的类中有不同的实现方式，称为多态。
class animal(): 
    def say(self):
        print("动物叫声")
class dog(animal):
    def say(self):
        print("汪汪汪")
class cat(animal):
    def say(self):
        print("喵喵喵")
dog = dog()
cat = cat()
dog.say()#调用dog类的say方法
cat.say()#调用cat类的say方法
```
