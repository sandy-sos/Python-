```python
#分支结构
if 5 > 3:
    print("5大于3") 

if 5 < 3:
    print("5小于3") 
elif 5 == 3:
    print("5等于3")
else:
    print("5不小于3")
```

```python
#与 and 或 or 非 not
a=100
print(a>0 and a<200) #True
print(a>0 or a<200) #True
print(not(a>0)) #False

```

```python
#嵌套结构 注意格式对齐
a=10
b=5

if a>0:
    print("a大于0")
    if b>0:
        print("b大于0")
    else:
        print("b不大于0")
else:
    print("a不大于0")
```

```python
num=1;
while num<5:
    print(num)
    num+=1
print("循环结束%d" % num)
```

```python
#1-10的和
a=1
sum=0
while a<=10:
    sum+=a
    a+=1
print("1-10的和为%d" % sum)
```

```python
#1-100偶数和
a=1
sum=0
while a<=100:
    if a%2==0:
        sum+=a
    a+=1
print("1-100偶数和为%d" % sum)

        
```

```python
#九九乘法表
i=1
while (i<10):
    j=1
    while j<=i:
        print("%d*%d=%d" % (i,j,i*j), end="\t")
        j+=1    
    print()
    i+=1        
```

```python
#输出*
a=1
while a<5:
    print("*"*a)#直接*a,代表重复a次
    a+=1
```

```python
#for循环 区间左闭右开   
for i in range(10): 
    print(i) #0-9

```

```python
# 打印九九乘法表
for i in range(1, 10):
    for j in range(1, i+1):
        #print("%d*%d=%d" % (i, j, i*j), end="\t")
        print(f"{i}*{j}={i*j}" ,end="\t")
    print()
    
```

```python
#自定义函数   
def add(a,b):
    return a+b

# 调用函数
res=add(1,1)
print(res)
```

```python
#匿名函数 函数名 = lambda 参数: 表达式 
sum = lambda a,b: a+b
sum(1,2)
```

```python
#递归
def fun(n):
    if(n==1):
        return 1
    return n*fun(n-1)

fun(5)
```
