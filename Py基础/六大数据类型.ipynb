```python
a=3.1415
print(type(a))#type()函数可以查看变量的类型
print("a的值是%.2f" % (a)) #%.2f可以保留两位小数
print("a的值是{:.1f}".format(a)) #{:.2f}可以保留两位小数
print(f"a的值是{a:.3f}") #f字符串可以保留两位小数
```

```python
#复数类型
a=1+2j
print(type(a))

b=complex(3,4) #complex()函数可以创建复数
print(type(b))

print(a) #复数的实部和虚部分别用a.real和a.imag来访问
print(a.real) #复数的实部
print(a.imag) #复数的虚部

print(b) 
print(b.real) #复数的实部和虚部都是浮点类型
print(b.imag)

print(a+b) #复数的加法
```

```python
a="abcd"
#基本语法squence[start:end:step]
print(a[-1]) #字符串的索引可以从右往左数，-1表示最后一个字符
print(a[1:3]) #字符串的切片可以得到子串，1:3表示从索引1到索引3（不包括3）的子串
print(a[::2]) #step表示步长，2表示每隔一个字符取一个字符
print(a[::-1]) #step为负数表示反向切片，可以得到字符串的反转
```

```python
list_a=[1,2,3,4,5,"my name is python",[1,2,3]]  #列表可以包含不同类型的元素
print(list_a[5]) #列表的索引和切片与字符串类似
print(list_a[5][0]) #列表中的字符串也可以使用索引来访问字符
print(list_a[-1])
print(list_a)
print(list_a[-7:-1]) #切片可以得到列表的子列表  
print(list_a[-7:]) 
```

```python
a=[1,2,3,4]
b=[1,2,3,6]
c=a+b  #列表的加法可以得到一个新的列表，包含两个列表的元素,但是不会去重
print(c)

c.sort() #sort()方法可以对列表进行排序，默认是升序
print(c)

c.sort(reverse=True) #reverse=True表示降序
print(c)
```

```python
d=list(range(1,5)) #range()函数可以生成一个整数序列，list()函数可以将其转换为列表
print(d)

d.append(5) #append()方法可以在列表末尾添加一个元素
print(d)

d.remove(1) #remove()方法可以删除列表中的一个元素，参数是要删除的元素的值
print(d)

d[0]=10 #列表是可变类型，可以修改列表中的元素
print(d)

print(max(d)) #max()函数可以得到列表中的最大值
```

```python
#元组类型,元素不可以修改 【】是列表 （）是元组
tuple_a=(1,1,1,1,1.2,"abc",[1,2,3]) #元组可以包含不同类型的元素
print(tuple_a[2]) #元组的索引和切片与字符串类似

b=tuple_a.count(1) #count()方法可以统计元组中某个元素出现的次数
print(b)

c=tuple_a.index(1.2) #index()方法可以返回元组中某个元素第一次出现的索引
print(c)
```

```python
#集合是无序且不重复的。数学概念的集合一致 {}是集合或者set() 【】是列表 （）是元组
#因为是无序的，所以不支持切片与索引操作
my_set={1,2,3,4,5,1,2,3} #集合可以包含不同类型的元素，但是不能包含可变类型的元素
print(my_set) #集合是无序的，重复的元素会被自动去重

my_set2=set([1,2,3,4,5,1,2,3]) #set()函数可以将列表转换为集合
print(my_set2)

#空集合只可使用set()函数创建，不能使用{}创建，因为{}表示空字典
empty_set=set() #空集合
print(empty_set)

a_set={1,2,3,4,5,5,5}
a_set.add(6) #add()方法可以向集合中添加一个元素
print(a_set)

a_set.remove(1) #remove()方法可以从集合中移除一个元素
print(a_set)

del a_set #del语句可以删除整个集合
# print(a_set) #删除后再访问会报错


a={1,2,3,4,5}
b={1,2}
c=a&b #&运算符可以得到两个集合的交集
print(c)

c=a|b #|运算符可以得到两个集合的并集
print(c)

d=a-b #-运算符可以得到两个集合的差集
print(d)



#字典类型  如果有两个相同的键，后面的值会覆盖前面的值
#字典是无序的键值对集合，键必须是不可变类型，值可以是任意类型
my_dict={"name":"python","age":30,"gender":"male"} 

print(my_dict["name"]) #字典的索引是键，返回的是值

my_dict["age"]=31 #可以通过键修改字典的值
print(my_dict)

my_dict["address"]="China" #可以通过键添加新的键值对
print(my_dict)

del my_dict["gender"] #可以通过键删除键值对
print(my_dict)

keys=my_dict.keys() #keys()方法可以返回字典的所有键
print(keys)

values=my_dict.values() #values()方法可以返回字典的所有值
print(values)
```
