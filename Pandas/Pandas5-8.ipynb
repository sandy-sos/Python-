```python
#pandas 合并concat

import pandas as pd
import numpy as np

#concatenate函数用于沿一个轴将多个数据结构连接起来。
# 默认情况下，concat沿着axis=0（行）进行连接。

df1 = pd.DataFrame(np.zeros((3, 4)), columns=['A', 'B', 'C', 'D'])
df2 = pd.DataFrame(np.ones((3, 4)), columns=['A', 'B', 'C', 'D'])
df3 = pd.DataFrame(np.ones((3, 4)) * 2, columns=['A', 'B', 'C', 'D'])
print("df1:")
print(df1)
print("df2:")
print(df2)
print("df3:")
print(df3)

print("end================================")
#上下合并 axis=0表示按行合并，axis=1表示按列合并#ignore_index=True表示重新索引 
result = pd.concat([df1, df2, df3],axis=0,ignore_index=True) 

print("上下合并结果:")
print(result)

```

```python
#jion函数用于将两个数据集按照某个键进行连接，类似于SQL中的JOIN操作。
# 默认情况下，join是按行索引进行连接的。

df1 =pd.DataFrame(np.ones((3,4))*0,columns=['A','B','C','D'],index=[1,2,3])
df2=pd.DataFrame(np.ones((3,4))*1,columns=['B','C','D','E'],index=[2,3,4])

print("df1:")
print(df1)
print("df2:")
print(df2)

#outer表示取并集，inner表示取交集
#没有数据的地方用NaN填充
res=pd.concat([df1,df2],join='outer',ignore_index=True) 
print("连接结果:")
print(res)

print("end================================")
res=pd.concat([df1,df2],join='inner',ignore_index=True)
print("连接结果:")#只链接共同拥有的列
print(res)

```

```python
print("左连接")
#join是管理列的，对行进行操作用axis=1
res=pd.concat([df1,df2],axis=1,join='outer' ) #按df1的索引连接
print("连接结果:")
print(res)
#axis	拼接方向	join 控制什么
#axis=0（默认）	上下拼（加行）	join 控制列怎么对齐
#axis=1	左右拼（加列）	join 控制行怎么对齐

print("添加行数据")
new_row = pd.DataFrame({'A': [7], 'B': [8]})
df1 = pd.concat([df1, new_row], ignore_index=True)
print("添加行后的数据:")
print(df1)


print("添加列数据")
new_col = pd.DataFrame({'E': [9, 10, 11, 12]})
df1 = pd.concat([df1, new_col], axis=1)
print("添加列后的数据:")
print(df1)



```

```python
#pandas merge函数用于将两个数据集按照某个键进行连接，
# 类似于SQL中的JOIN操作。

import pandas as pd
import numpy as np

#创建两个数据集
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})       

print("left:")
print(left)
print("right:")
print(right)
#按key列进行连接，默认是inner连接
result = pd.merge(left, right, on='key')
print("连接结果:")
print(result)


```

```python
#how=['inner', 'outer', 'left', 'right']是连接方式，默认为'inner'。
#consider two keys
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})    

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

print("left:")
print(left)
print("right:")
print(right)

print("连接结果:")
#按key1和key2列进行连接，默认是inner连接
result = pd.merge(left, right, on=['key1', 'key2'], how='inner')
print(result)

print("end===============================")
#按key1和key2列进行连接，left连接,left数据不变，right 数据不匹配的地方用NaN填充
result = pd.merge(left, right, on=['key1', 'key2'], how='left')
print(result)
```

```python
#indicator参数用于指示每一行的来源，取值为True或False，默认为False。

df1=pd.DataFrame({'col1':[0,1],'col_left':['a','b']})
df2=pd.DataFrame({'col1':[1,2,2],'col_right':[2,2,2]})

print("df1:")
print(df1)
print("df2:")
print(df2)

#按col1列进行连接，indicator=True表示显示每一行的来源
result = pd.merge(df1, df2, on='col1', how='outer', indicator=True)
print("连接结果:")
print(result)
```

```python
#merge by index
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C1', 'C2'],
                      'D': ['D0', 'D1', 'D2']},
                      index=['K0', 'K2', 'K3'])
print("left:")
print(left)
print("right:")
print(right)

#按索引进行连接
result = pd.merge(left, right, left_index=True, right_index=True, how='outer', indicator=True)
print("连接结果:")
print(result)
```

```python
#handle overlapping column names
boys = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
girls = pd.DataFrame({'key': ['K0', 'K0', 'K2'], 'A': ['B0', 'B1', 'B2']})
print("boys:")
print(boys)
print("girls:")
print(girls)

#按key列进行连接，suffixes参数用于处理重叠的列名
result = pd.merge(boys, girls, on='key', how='outer', suffixes=('_boys', '_girls'))
print("连接结果:")
print(result)


```
