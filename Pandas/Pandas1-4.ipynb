```python
#pandas意义和功能：
#pandas是一个开源的Python数据分析库，提供高性能、易用的数据结构和数据分析工具。
#主要功能包括：
#1. 数据清洗和准备
#2. 数据操作和转换
#3. 数据分析和统计
#4. 数据可视化


import pandas as pd
import numpy as np

s = pd.Series([1, 2, 3,np.nan, 4, 5])
print(s)
print("end=============================================")

dates = pd.date_range('20260101', periods=6)
print(dates)

print("DataFrame以ABCD为列名，dates为行名，6行4列的随机数矩阵")
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=['A','B','C','D'])
print(df)

print("end=============================================")
print("随机的DataFrame，3行4列")
de1=pd.DataFrame(np.arange(12).reshape(3,4))
print(de1)
```

```python
print("end=============================================")
print("Timestamp:时间数据，Series:一维数据，DataFrame:二维数据," 
"array:多维数据,Categorical:分类数据")
df2=pd.DataFrame({
    'A': 1.,
    'B': pd.Timestamp('20230101'),
    'C': pd.Series(range(4), index=list(range(4)), dtype='float32'),
    'D': np.array([3] * 4, dtype='int32'),
    'E': pd.Categorical(["test", "train", "test", "train"]),
    'F': 'foo'
})
print(df2)
print("df2.dtypes:查看每一列的数据类型")
print(df2.dtypes)
print("df2.index:查看行索引")
print(df2.index)
print("df2.columns:查看列索引")
print(df2.columns)
print("df2.values:查看数据")
print(df2.values)

```

```python
#count	计数	非空值个数
#mean	均值	平均值
#std	标准差	数据离散程度（0=无波动，越大越分散）
#min	最小值	
#25%	第一四分位数	25%的数据比这个值小
#50%	中位数	50%的数据比这个值小（= median）
#75%	第三四分位数	75%的数据比这个值小
#max	最大值	
print("df2.describe():查看数据的统计信息")
print(df2.describe())#只计算数字

print("end==============================================")
print("df2.T:转置")
print(df2.T)

```

```python
print("#按列索引降序排列,axis=1表示按列索引排序，ascending=False表示降序排列")
print(df2.sort_index(axis=1, ascending=False))

print("end==============================================")
print("axis=0表示按行索引排序，ascending=False表示降序排列")
print(df2.sort_index(axis=0, ascending=False))

print("end==============================================")
print("按列C的值降序排列")
print(df2.sort_values(by='C', ascending=False))




```

```python
#pandas 选择数据

dates=pd.date_range('20260101', periods=6)
df=pd.DataFrame(np.arange(24).reshape((6,4)), index=dates, columns=['A','B','C','D'])
print(df)

print("end==============================================")

# 选择单列
print("df['A']:选择单列")
print(df['A'],'\n',df.A)#两个格式一致

```

```python
print("前三行数据")
#选择行，前者是按位置选择，后者是按索引选择
print(df[0:3],'\n',df['20260101':'20260103'])
print("end==============================================")

#selct by label:loc
print("选择特定行的数据")
print(df.loc['20260101'])#变成竖的了

print("end==============================================")
print("选择特定列的数据")
print(df.loc[:, ['A','B']])


```

```python
print("特定的行和列")
print(df.loc['20260102':'20260104', ['A','B']])

print("end==============================================")
#select by position:iloc
print("按位置选择数据:第三行，第一位")
print(df.iloc[2, 0])

print("end==============================================")
print("按位置选择数据:第二到三行，第一到第二列")
print(df.iloc[1:3, 0:2])#这里是冒号

print("end==============================================")
print("挑选特定的行，列")
print(df.iloc[[1,2,4],[0,2]])#这里是逗号

print("end==============================================")
#mixed selection:ix
print("混合选择数据,结合标签和数字")
print(df.iloc[:3][['A', 'C']])


```

```python

#Boolean indexing

print("df[df.A > 8]:选择A列大于8的行")
print(df[df.A > 8])#只用A对比，但显示一整行

print("end==============================================")
```

```python
#pandas设计值
print(df)
print("end===================================")

print("df.iloc[2,2]=1111:将第三行第三列的值改为1111")
df.iloc[2,2]=1111
print(df)
print("end===================================")

print("df.loc['20260101','B']=2222:将20260101行,B列的值改为2222")
df.loc['20260101','B']=2222
print(df)

```

```python


print("df[df.A > 4] = 0:将A列大于4的值改为0")
df[df.A > 4] = 0
print(df)
print("end===================================")


df.loc[df.A== 0,'A'] = 5
print("df.loc[df.A == 0, 'A'] = 5:将A列中值为0的改为5")
print(df)

print("end===================================")

df['F']="A"
print("df['F']='A':新增一列F，值为A")
print(df)
```

```python

df['E']=pd.Series([1,2,3,4,5,6],index=dates)#上述写过，直接用dates，重新赋值也可以
print("df['E']=pd.Series([1,2,3,4,5,6],index=dates):新增一列E，值为1-6")
print(df)
```

```python
#pandas 处理丢失数据
print("原数据")
print(df)
print("end===================================")

print("df.dropna():删除所有含有NaN的列")
df.iloc[0, 0] = np.nan 
print(df.dropna(how='any', axis=1))#删除所有含有NaN的列
#how='any'（默认）	该行/列有至少一个 NaN 就删
#how='all'	该行/列全部都是 NaN 才删

print("end===================================")
print("df.fillna(value=0):将所有NaN的值替换为0")
print(df.fillna(value=0))#将所有NaN的值替换为0


```

```python
print("判断是否为NaN，返回True或False")
print(df.isnull())#判断是否为NaN，返回True或False
print("end===================================")

print("df.isnull().sum():统计每一列中NaN的个数")
print(df.isnull().sum())#表格太大使用

print("end===================================")
print("np.any(df.isnull()):判断是否有NaN，返回True或False")
print(np.any(df.isnull()) )#判断是否有NaN，返回True或False



```
