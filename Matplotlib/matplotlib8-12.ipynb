```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

#annotation
x=np.linspace(-3,3,100)
y1=2*x+1

plt.figure(num=22, figsize=(8, 5))
plt.plot(x, y1)
plt.plot(x,y1, color='red', linewidth=1.0, linestyle='--')

ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


x0=1
y0=2*x0+1
plt.scatter(x0,y0,s=50,c='b')#s是size,c=color
plt.plot([x0,x0],[y0,0],'k--',lw=2.5)
#k = 黑色, -- = 虚线,	线宽 2.5
#[x0, x0]	x 坐标不变（两个点都是 x0），画竖线
#[y0, 0]	y 从 y0 到 0


#方法1
plt.annotate(r'$2x+1=%s$' %y0,xy=(x0,y0),xycoords='data',xytext=(+30,-30),
             textcoords='offset points',fontsize=16,arrowprops=dict(arrowstyle='->',
             connectionstyle='arc3,rad=.2'))
#xy	箭头指向的目标点
#textcoords	定义xytext的坐标系统 （像素偏移x+30，y-30）
#arrowprops	箭头样式字典 
#connectionstyle='arc3,rad=.2' 弧度

#文字放在 xytext
#                   ▲
#                   │ 箭头从文字指向 xy
#                   │
#                   ●目标点 xy


#方法2
#plt.text(-3.7,3,r'$this\ is\ the\ some\ text.\ mu\\ sigma_i\\alpha_t$',
 #        fontdict={'size':16,'color':'r'})
# 带希腊字母（用LaTeX，修正反斜杠）
plt.text(-3.7, 3, r'$this\ is\ the\ some\ text.\mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 16, 'color': 'r'})
plt.show()
```

```python
#tick能见度:让坐标数值清晰可见

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y = 0.1*x

plt.figure()
plt.plot(x, y, linewidth=10, zorder=1)     
# set zorder for ordering the plot in plt 2.0.2 or higher
#z 值越大，画在越上面（越靠前），不会被 z 值小的元素遮盖。
#zorder=3   → 最顶层（最前）
#zorder=2   → 中间层
#zorder=1   → 底层（最后面）
#zorder=0   → 默认层

plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))


for lable in ax.get_xticklabels()+ax.get_yticklabels():#选择X轴和Y轴
    lable.set_fontsize(12)
    lable.set_bbox(dict(facecolor='white',edgecolor='None',alpha=0.7))
#facecolor	框的背景填充色，'white' 白色
#edgecolor	边框颜色，'None' = 不画边框
#alpha	透明度，0.7 让背景半透明，能隐约看到背后的图形


plt.show()
```

```python
#scatters 散点图

n=1024
X=np.random.normal(0,1,n)
Y=np.random.normal(0,1,n)

T=np.arctan2(Y,X) #颜色
#点与原点夹角 0° → 一种颜色
#点与原点夹角 90° → 另一种颜色
#点与原点夹角 180° → 又一种颜色
#一圈下来颜色渐变，像一个彩色轮盘

plt.scatter(X,Y,s=75,c=T,alpha=0.5) 
#s=size,c=color,alpha透明度

plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))

plt.xticks(())
plt.yticks(())

plt.show()

```

```python
#柱状图

n=12
X=np.arange(n)
Y1=(1-X/float(n)*np.random.uniform(0.5,1.0,n))
Y2=(1-X/float(n)*np.random.uniform(0.5,1.0,n))

plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.04,y+0.05,'%.2f'%y,ha='center',va='bottom')
    
    
for x,y in zip(X,Y2):
    plt.text(x+0.04,-y-0.05,'-%.2f'%y,ha='center',va='top')
#x+0.04, y+0.05	文字位置：在点 (x, y) 的右上方偏移一点
#ha='center'	水平对齐方式 horizontal alignment
#va='bottom'	垂直对齐方式 vertical alignment


plt.xlim(-0.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())

plt.show()
```

```python
#contours等高线图

import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
#网格 X,Y = np.meshgrid(x, y)

# use plt.contourf to filling contours
# X, Y and F(X,Y)==Z
plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
#cmap=colormap 把数值映射到颜色的规则——数值从小到大，颜色从暗到亮渐变。
#还有cool,gray,Blues


# use plt.contour to add contour lines
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
#数字8代表把等高线分成几份

# adding label
plt.clabel(C, inline=True, fontsize=10)#画在等高线里面

plt.xticks(())
plt.yticks(())
plt.show()
```
