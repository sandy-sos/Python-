```python
%matplotlib inline 
#内置窗口

import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(-1,1,100)
y=2*x+1
plt.plot(x,y)
plt.show()

```

```python
%matplotlib qt
#独立窗口
#figure图像:先定义一个figure，然后在figure上绘制图像，
# figure可以有多个，figure编号从1开始，默认编号为1
x=np.linspace(-1,1,100)
y1=2*x+1
y2=x**2
plt.figure()#第一个figure
plt.plot(x, y1)

plt.figure(num=22, figsize=(8, 5))#第二个figure
#num=22表示图像编号，figsize表示图像大小
plt.plot(x, y2)
plt.plot(x,y1, color='red', linewidth=1.0, linestyle='--')
#linewidth表示线宽，linestyle表示线型

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#坐标轴的设置1

x=np.linspace(-1,1,100)
y1=2*x+1
y2=x**2

plt.plot(x, y2)
plt.plot(x,y1, color='red', linewidth=1.0, linestyle='--')

plt.xlim((-1,2))#设置x轴的范围
plt.ylim((-2,3))#设置y轴的范围

plt.xlabel('I am x')#设置x轴的名称
plt.ylabel('I am y')#设置y轴的名称

new_ticks=np.linspace(-1,2,5)#设置x轴的刻度
plt.xticks(new_ticks)#设置x轴的刻度

plt.yticks([-2,-1,0,1,2,3],['really bad','bad','normal','good','really good','excellent'])#设置y轴的刻度


plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#坐标轴的设置2

#gca='get current axis'获取当前的坐标轴信息
ax=plt.gca()
ax.spines['right'].set_color('none')#设置右边的坐标轴为无色
ax.spines['top'].set_color('none')#设置上边的坐标轴

ax.xaxis.set_ticks_position('bottom')#设置x轴的刻度在下方
ax.yaxis.set_ticks_position('left')#设置y轴的刻度在左边
ax.spines['bottom'].set_position(('data',0))#设置x轴的位置在
ax.spines['left'].set_position(('data',0))#设置y轴的位置在
#不仅可以设置在data位置，还可以设置在outward和axes位置

x=np.linspace(-1,1,100)
y1=2*x+1
y2=x**2

plt.plot(x, y2)
plt.plot(x,y1, color='red', linewidth=1.0, linestyle='--')




plt.show()
```

```python

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#Legend图例的设置

x=np.linspace(-1,1,100)
y1=2*x+1
y2=x**2


plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks=np.linspace(-1,2,5)
plt.xticks(new_ticks)
plt.yticks([-2,-1,0,1,2,3],['really bad','bad','normal','good','really good','excellent'])#设置y轴的刻度

y2,=plt.plot(x, y2, label='up')#label设置图例的名称
y1,=plt.plot(x,y1, color='red', linewidth=1.0, linestyle='--', label='down')#设置图例的名称
plt.legend(loc='best',labels=['up111','down111'],handles=[y2,y1])
#best找一个适合的位置放图例的位置
#还有：upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center
#handles指定图例的对象，labels指定图例的名称


plt.show()
```
